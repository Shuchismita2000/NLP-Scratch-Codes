```python
import nltk as nlp 
from nltk.corpus import brown
import re 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
import collections
```

```python
bad_chars=['.',',',';',':','!','?','@','#','$','%','^','*','(',')','``']
x=[w for w in brown.words(categories='adventure') if w not in bad_chars]
word_frequencies = collections.Counter(x)
top_word_frequencies = word_frequencies.most_common()
table1=pd.DataFrame(top_word_frequencies,columns=['words','frequency'])
table1["rank"]=table1.index+1
k=table1["frequency"][0]*table1["rank"][0]
table1['pred']=k/table1["rank"]
table2=table1.head(20)
print('According to the rank (1) ,suppose the value of k is',k)
table2
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.lineplot(x="words",y='frequency',label='original',data=table2)
sns.lineplot(x="words",y='pred',label='Pred ',data=table2)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
