```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import brown
```

```python
q=brown.tagged_sents(tagset='universal')
q1=[]
for i in range(len(q)):
    q1.append(q[i][0][1])
counts3=dict()
for i in range(len(q1)):
    if q1[i] in counts3:
        counts3[q1[i]]+= 1
    else:
        counts3[q1[i]]=1
df2=[]
df2.append(counts3)
df2=pd.DataFrame(df2)
df2=df2/len(q)
df2=df2.T
df2.rename(columns={0:"Pi Probability"}, inplace=True)
df2
```

```python
p=brown.tagged_words(tagset='universal')
wlist=[]
y=[]
for i in p:
    wlist.append(i)
    if(len(wlist)==2):
        y.append(wlist)
        wlist=wlist[1:]  
h1=[]
for i in range(len(y)):
    h1.append((y[i][0][1],y[i][1][1]))
result={}
for i in h1:
    if i[0] in result:
        result[i[0]]=(*result[i[0]],*i[1:])
    else :
        result[i[0]] = i
r=list(result.values())
n=[]
n1=[]
for i in range(len(r)):
    n.append(r[i][0])
    l1=list(r[i])
    l1.pop(0)
    n1.append(l1)
counts2=dict()
df1=[]
for i in range(len(n1)):
    for j in n1[i] :
        if j in counts2:
            counts2[j]+= 1
        else:
            counts2[j]=1
    df1.append(counts2)
    counts2=dict()
df1=pd.DataFrame(df1)
df1=df1.replace(np.nan,0)

df1['sum']=df1['NOUN']+df1['VERB']+df1['DET']+df1['ADJ']+df1['PRON']+df1['NUM']+df1['ADV']+df1['ADP']+df1['.']+df1['X']+df1['PRT']+df1['CONJ']
df_new = df1.loc[:,"NOUN":"CONJ"].div(df1["sum"], axis=0)
df_new['#']=n
df_new.set_index('#')
```
